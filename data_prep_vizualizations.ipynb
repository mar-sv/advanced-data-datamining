```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Mount Google Drive
drive.mount('/content/drive')
```

```python


# Load the train and test datasets
train_df = pd.read_csv('/content/drive/My Drive/D7043E_Project/train_dataset.csv')
test_df = pd.read_csv('/content/drive/My Drive/D7043E_Project/test_dataset.csv')

# Drop the 'Unnamed: 0' column (row number)
train_df.drop(columns=['Unnamed: 0'], errors='ignore', inplace=True)
test_df.drop(columns=['Unnamed: 0'], errors='ignore', inplace=True)

```

```python
# Bar Chart for Train and Test Data Distribution by Percentage
train_rows = train_df.shape[0]
test_rows = test_df.shape[0]
total_rows = train_rows + test_rows

# Show percentage of rows used in each dataset
train_percentage = (train_rows / total_rows) * 100
test_percentage = (test_rows / total_rows) * 100

# Bar chart to visualize the distribution
fig, ax = plt.subplots(figsize=(7, 5))
fig.patch.set_facecolor('#1E3A5F')  # Set the overall background color
ax.set_facecolor('#1E3A5F')  # Set the background color of the plot
bars = ax.bar(['Train Dataset', 'Test Dataset'], [train_percentage, test_percentage], color=['green', 'red'])

# Data labels (percentage and total number of rows)
for bar, total in zip(bars, [train_rows, test_rows]):
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 1, f'{yval:.2f}%', ha='center', va='bottom', color='white')
    ax.text(bar.get_x() + bar.get_width() / 2, yval - 10, f'Total: {total}', ha='center', va='bottom', color='white')

# Vizulization style
ax.set_title('Train vs Test Dataset Distribution by Percentage', color='white')
ax.set_ylabel('Percentage distribution (%)', color='white')
ax.set_xlabel('Dataset', color='white')
ax.set_ylim(0, 100)
ax.tick_params(colors='white')

plt.show()
```
